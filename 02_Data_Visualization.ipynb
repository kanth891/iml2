# Experiment 2: Data Visualization Techniques
**Objective:** Apply data visualization techniques using Python libraries Matplotlib and Seaborn on the Iris dataset.

```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris Dataset
df = sns.load_dataset("iris")
print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# 1. Line Plot (Matplotlib) – Trend of Sepal Length
plt.figure(figsize=(8, 4))
plt.plot(df['sepal_length'], color='steelblue')
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram (Matplotlib) – Distribution of Petal Length
plt.figure(figsize=(8, 4))
plt.hist(df['petal_length'], bins=20, color='orange', edgecolor='black')
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# 3. Bar Plot (Seaborn) – Count of Each Species
plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df, palette='Set2')
plt.title("Count of Each Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# 4. Scatter Plot (Seaborn) – Relationship Between Sepal & Petal Length
plt.figure(figsize=(7, 5))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, palette='Set1')
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.tight_layout()
plt.show()
```

```python
# 5. Box Plot (Seaborn) – Spread and Outliers in Sepal Width
plt.figure(figsize=(7, 5))
sns.boxplot(x='species', y='sepal_width', data=df, palette='Pastel1')
plt.title("Sepal Width by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.tight_layout()
plt.show()
```

```python
# 6. Heatmap (Seaborn) – Correlation Between Features
plt.figure(figsize=(7, 5))
correlation = df.drop('species', axis=1).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
```
